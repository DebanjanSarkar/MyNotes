# MATPLOTLIB:-
----------------
- Python library to plot 2D graphs and charts.
- It is mainly used for data visualization.
- Using it we can generate **plots, histograms, power spectra, bar graphs, pie charts, error charts**, etc...

```python
import matplotlib as mtlb
from matplotlib import pyplot as plt

import numpy as np
```

## Plotting points in a xy - plane:
## `pyplot.plot( <x_values> , <y_values> , label=<name_to_denote_curve> )` and `pyplot.show()`

```python
# We want to plot these points, and a graph using them:- (1,21), (2,15), (3,17), (4,35), (5,108)
# We separate their x-values in a list, and y-values in another list, and pass then to pyplot.plot() function parameters.

x_val = [1, 2, 3, 4, 5]
y_val = [21, 15, 17, 35, 98]

plt.plot( x_val, y_val )

plt.xlabel("Tests")
plt.ylabel("Marks")
plt.title("Test Scores")

plt.show() # Without this line, the plot will not be displayed.
```

### Plotting multiple graphs in a single canvas:
---------------------------------------------------
- For each call of `pyplot.plot()` function, a new curve is being created in the canvas.
- Thus, for each separate curve we want to create, we want to call `pyplot.plot()` function that many times, with the different values of x and y.
- To identify each curve, we need to set the `label` keyword argument for each plot, and then call the `pyplot.legend()` function before displaying the plot.

```python
sem = [1, 2, 3, 4, 5, 6, 7, 8]
gpa_deb = [8.84, 7.33, 4.25, 9.95, 6.43, 9.33, 8.5, 8.65]
gpa_jeet = [4.65, 5.24, 5.95, 6.33, 6.8, 7.5, 7.8, 8.9]

plt.plot( sem, gpa_deb, label="Deb" )
plt.plot( sem, gpa_jeet, label="Jeet" )

plt.xlabel("B.Tech Semester")
plt.ylabel("Grade Point Average (GPA)")
plt.title("Semester-wise GPA of Students")
plt.legend()
plt.show()
```

## Bar Graphs :

```python
sem = [1, 2, 3, 4, 5, 6, 7, 8]
gpa_deb = [8.84, 7.33, 4.25, 9.95, 6.43, 9.33, 8.5, 8.65]

plt.bar( sem, gpa_deb, label="Deb" )
plt.xlabel("BTech Semester")
plt.ylabel("GPA")
plt.legend()
plt.title("Semesterwise Marks of Deb")
plt.show()
```

### Changing width of each bar using `width` keyword argument.
- `width=1` is normal default width.
- `width=0.2` will make the width of each bar (1/5) times of default width.
- `width=0.5` will make the width of each bar (1/2) times of default width.
- `width=2` will make the width of each bar 2 times of default width.
... so on.

```python
sem = [1, 2, 3, 4, 5, 6, 7, 8]
gpa_deb = [8.84, 7.33, 4.25, 9.95, 6.43, 9.33, 8.5, 8.65]

plt.bar( sem, gpa_deb, label="Deb", width=0.2 )
plt.xlabel("BTech Semester")
plt.ylabel("GPA")
plt.legend()
plt.title("Semesterwise Marks of Deb")
plt.show()
```

## HISTOGRAMS:-

- Histogram is a bar graph that visualizes data distribution.
- Here, we visualize number of occurrences, i.e., the things that occur lies in x-axis, and number of occurences lies in the y-axis.
    * Ex:- if we 20 test scores of a student, and want to see how many lies between 0-30, 30-50, 50-70, 70-90, 90-100, for that we can use histogram. 
- Thus, we do not pass normal x and y axis values to it. We pass the dataset which contains the data that occurs, and number (limits/boundaries) of sections.
- Histograms can be of many type. Ex- bar histogram, step histogram.
- Number of sections == number of bars in bar histogram. 

#### `pyplot.hist( <data_to_visualize> , <section_as_a_list_of_integers> , histtype="bar/etc..." , rwidth = <relative_width(number)> )`

##### Q. Plot the distribution / number of matches in which runs scored is betwen 0-25, 25-50, 50-75, 75-100, 100-125.
Runs = 0, 41, 57, 67, 32, 108, 44, 9, 17, 81, 99, 91, 64, 3, 27, 55, 50 

```python
runs_per_innings = [ 0, 41, 57, 67, 32, 108, 44, 9, 17, 81, 99, 91, 64, 3, 27, 55, 50 ]
sections = [0,25,50,75,100,125]   # We made sections for 0-25, 25-50, 50-75, 75-100, 100-125

plt.hist( runs_per_innings , sections , histtype="bar" , rwidth=1 )
plt.xlabel("Runs scored")
plt.ylabel("Number of Innings")
plt.title("Runs scored per innings v/s number of innings")
plt.grid()
plt.show()
```

## STACK PLOT :-
- A Stack Plot visualises the entire dataset with easy visualization of how each part (each individual data in the dataset) makes up the whole.
- Each data in the data set is a contituent of the stack plot, and each such constituent (data) is stacked on top of each other.

### `pyplot.stack( <x_axis_data> , <dataset_1> , <dataset_2> ,..., <dataset_n> , labels = <list_containing_labels> )`


Q. There are 7 days in a week, and 24 hours in each day. You have done following activities each day, with number of hours spent in each is respresented as each number in the list:

sleep=[8,7,6,7,5,4,9] <br>
study=[4,2,6,1,3,5,5] <br>
workout=[1,2,1,1,2,3,1] <br>
phone=[5,3,2,4,1,3,4] <br>
other=[6,10,9,11,13,9,5] <br>

Make a scatter plot showing the distribution, for each day of a week.

```python
days = ["sun", "mon", "tue", "wed", "thurs", "fri", "sat"]

sleep=[8,7,6,7,5,4,9]
study=[4,2,6,1,3,5,5]
workout=[1,2,1,1,2,3,1]
phone=[5,3,2,4,1,3,4]
other=[6,10,9,11,13,9,5]

labels_list = ["Sleep", "Study", "Workout", "Used Phone", "Other"]

plt.stackplot( days , sleep,study,workout,phone,other , labels = labels_list )
plt.xlabel("WEEKDAY")
plt.ylabel("Number of hours spent")
plt.title("Hours spent in each activity per day in a week")
plt.legend()
plt.grid()

plt.show()
```

## PIE CHART:-

### `pyplot.pie( <list_of_values_constituting_each slice>, labels = <list_of_labels_of_each_slice> )`


Optional keyword arguments:
- `explode = (.. , .. , .. )` : Takes a tuple, of float numbers, from 0 to positives. Tuple size must be same as the number of slices. Brings out that slice whose corresponding value in the tuple is non-zero.
- `autopct = '%<number_of_digits_before_decimal>.<number_of_digits_after_decimal>f'` : To show the percent of share each slice takes up of the whole.

```python
# days = ["sun", "mon", "tue", "wed", "thurs", "fri", "sat"]

sleep=np.array([8,7,6,7,5,4,9])
study=np.array([4,2,6,1,3,5,5])
workout=np.array([1,2,1,1,2,3,1])
phone=np.array([5,3,2,4,1,3,4])
other=np.array([6,10,9,11,13,9,5])

avg_hours_spent = [ sleep.mean(), study.mean(), workout.mean(), phone.mean(), other.mean() ]

labels_list = ["Sleep", "Study", "Workout", "Used Phone", "Other"]

plt.pie( avg_hours_spent , labels = labels_list )
plt.title("Average Hours spent in each activity per day in a week")
# plt.legend()

plt.show()
```

## Using File Data to Draw plots:-
- Here, we load the data from file into Python data structures, and use that to plot the graphs.

```python
names = []
marks = []

with open("./data/02_student.csv", "r") as student_file:
    # Uncomment this below line if the file contains header row, containing column names.
    # Reading one line will move the pointer to the next line, from wheree the actual data starts.
    student_file.readline()
    
    line = student_file.readline()
    while line:
        # print(line.strip())

        # Whole line is a single string. Splliting it based on "," to get individual data.
        details = line.strip().split(",")
        names.append( details[1].split()[0] )
        marks.append( int(details[3]) )

        #Read next line
        line = student_file.readline()

print( names, marks, sep="\n\n" )
```

### Increasing the size of Plot canvas

```python
plt.rcParams["figure.figsize"] = (18, 10)
```

```python
plt.bar( names, marks, width = 0.5, color="#00AA00" )

# Rotate the x-axis value labels (names of students) by 60 degrees
plt.xticks(rotation=60)

plt.xlabel("Names of Students")
plt.ylabel("Marks Obtained")
plt.title("Student Results")
plt.grid()

plt.show()
```
